### Conceptual model

#### Step 1
Using the `PrecipitationDistribution` component in Landlab we will generate a single storm that is uniform in space, but varies in rainfall flux over the duration of the storm. This will be done on a simple catchment (`hugo_site.asc`). Rainfall flux will contribute to `surface_water_depth` which will be fed into `soil_water_infiltration__depth`.

#### Step 2
With the `SoilInfiltrationGreenAmpt`, we will calculate the `soil_water_infiltration__depth` and the new `surface_water__depth`. 

#### Step 3
This `surface_water__depth` term is also used in `OverlandFlow`! So now we can run the `OverlandFlow` to determine where (and how much) the water goes.

### Define parameters and initialize PrecipitationDistribution

```python
## Define parameters for rainfall generator 
mean_duration = 6 ## hours, storm duration
mean_duration2 = 10 

mean_inter_duration = 4 ## hour,  duration between storms
mean_inter_duration2 = 10

mean_depth = 20   ## mm
mean_depth2 = 10

total_time  = 24*30 ## 24 hours * 30 days
delta_t = 1; ## hour
np.random.seed(np.arange(10))

total_time2  = 24*30 ## 24 hours * 30 days
delta_t2 = 1; ## hour
np.random.seed(np.arange(10))

# Initialize generator
precip = PrecipitationDistribution(mg, mean_storm_duration=mean_duration, 
                                   mean_interstorm_duration=mean_inter_duration, 
                                   mean_storm_depth=mean_depth, total_t=total_time,delta_t=delta_t)
n = random.randint(1,101) ## random seed number 
precip.seed_generator(seedval=n)

```

```python
#Comparison event
precip2 = PrecipitationDistribution(mg2, mean_storm_duration=mean_duration2, 
                                  mean_interstorm_duration=mean_inter_duration2, 
                                  mean_storm_depth=mean_depth2, total_t=total_time2,delta_t=delta_t2)
n2 = random.randint(1,101) ## random seed number 
precip2.seed_generator(seedval=n2)
```

## Generate rainfall series

```python
## Iterating on all of the storms and save rainfall data

# Arrays for saving the outputs.
storm_dts = []
interstorm_dts = []
intensities = []
depths  = [] 
total_time=0
time_vec =[0]
rain_vec = [0]
for (storm_dt, interstorm_dt) in precip.yield_storms():   #Each iteration has a storm duration and interstorm duration
        ## Save rainfall data                             #This loop will go over all the storms. precip.yield_storms(): 

    storm_dts.append(np.array(storm_dt)) ## hours
    interstorm_dts.append(interstorm_dt) ## hours
    intensities.append(mg.at_grid['rainfall__flux']) ## mm/h
    depths.append(mg.at_grid['rainfall__flux']*storm_dt)
    
    ## Creating rainfall time-series
    total_time += 1
    time_vec.append(total_time)
    rain_vec.append(intensities[-1])
    total_time += storm_dt
    time_vec.append(total_time)
    rain_vec.append(intensities[-1])
    total_time += 1
    time_vec.append(total_time)
    rain_vec.append(0)
    total_time += interstorm_dt
    time_vec.append(total_time)
    rain_vec.append(0)

time_vec = np.asanyarray(time_vec)*60 ##in minutes   #Full time series will all storms
rain_vec = np.asanyarray(rain_vec)                   #Rain intensities for full series (mm/hr)
```

```python
## Iterating on all of the storms and save rainfall data *Second event*

# Arrays for saving the outputs.
storm_dts2 = []
interstorm_dts2 = []
intensities2 = []
depths2  = [] 
total_time2=0
time_vec2 =[0]
rain_vec2 = [0]
for (storm_dt, interstorm_dt) in precip2.yield_storms():   #Each iteration has a storm duration and interstorm duration
        ## Save rainfall data                             #This loop will go over all the storms. precip.yield_storms(): 

    storm_dts2.append(np.array(storm_dt)) ## hours
    interstorm_dts2.append(interstorm_dt) ## hours
    intensities2.append(mg2.at_grid['rainfall__flux']) ## mm/h
    depths2.append(mg2.at_grid['rainfall__flux']*storm_dt)
    
    ## Creating rainfall time-series
    total_time2 += 1
    time_vec2.append(total_time2)
    rain_vec2.append(intensities2[-1])
    total_time2 += storm_dt
    time_vec2.append(total_time2)
    rain_vec2.append(intensities2[-1])
    total_time2 += 1
    time_vec2.append(total_time2)
    rain_vec2.append(0)
    total_time2 += interstorm_dt
    time_vec2.append(total_time2)
    rain_vec2.append(0)

time_vec2 = np.asanyarray(time_vec2)*60 ##in minutes   #Full time series will all storms
rain_vec2 = np.asanyarray(rain_vec2)                   #Rain intensities for full series (mm/hr)
```

```python
fig, ax = plt.subplots(1,2, figsize = (24,10))

## Plotting the full rainfall series
number_of_storms = str(len(storm_dts))
ax[0].plot(time_vec,rain_vec,linewidth=2)   
ax[0].set_xlabel('Time [min]',fontsize=20)
ax[0].set_ylabel('Rainfall intensity [mm $h^{-1}]$',fontsize=18)
ax[0].set_ylim([0,15])
ax[0].set_title('Rainfall Intensity',fontsize=24)
ax[0].text(10000, 10,'Number of storms: ' + number_of_storms, horizontalalignment='center', verticalalignment='center', fontsize =20)

## Plotting rainstorm depth CDF-plot

hist, bin_edges = np.histogram(depths) # This part was giving an error?, normed=True)
b = [0]
b.extend(np.cumsum(hist)/np.sum(hist))
ax[1].plot(bin_edges,b,label='Scenario #1',linewidth=2,color='firebrick')
ax[1].set_xlabel('Storm depth [m]',fontsize=18)
ax[1].set_ylabel('F(x)',fontsize=18)
ax[1].set_ylim([0,1.2])
ax[1].set_title('Storm Depth',fontsize=24)
ax[1].legend()


fig2, ax = plt.subplots(1,2, figsize = (24,10))

## Plotting the full rainfall series
number_of_storms2 = str(len(storm_dts2))
ax[0].plot(time_vec2,rain_vec2,linewidth=2)   
ax[0].set_xlabel('Time [min]',fontsize=20)
ax[0].set_ylabel('Rainfall intensity [mm $h^{-1}]$',fontsize=18)
ax[0].set_ylim([0,15])
ax[0].set_title('Rainfall Intensity',fontsize=24)
ax[0].text(10000, 10,'Number of storms: ' + number_of_storms2, horizontalalignment='center', verticalalignment='center', fontsize =20)

## Plotting rainstorm depth CDF-plot

hist2, bin_edges = np.histogram(depths2) # This part was giving an error?, normed=True)
b = [0]
b.extend(np.cumsum(hist2)/np.sum(hist2))
ax[1].plot(bin_edges,b,label='Scenario #1',linewidth=2,color='firebrick')
ax[1].set_xlabel('Storm depth [m]',fontsize=18)
ax[1].set_ylabel('F(x)',fontsize=18)
ax[1].set_ylim([0,1.2])
ax[1].set_title('Storm Depth',fontsize=24)
ax[1].legend()


# Let's create two months with contrasting rain series/distributions compare outputs --> subsequent hydrographs 
```

## Runoff and the infiltration

In the model, at each time step and for each cell, the rain accumulates the water column above ground each cell. Some of the water infiltrates with the SoilInfiltrationGreenAmpt and the remainder is moved through the basin drainage with the OverlandFlow component, 

### Overlandlandflow component
The elevation grid controls the overland flow component. Slope and basin drainage size can have strong impact on the stream hydrograph.

### Soil Infiltration component
The infiltration of water in the soil can be described by $q$, which is the velocity at which the water is moving through the soil. $q$ is usually in $m/s$ and is controlled by the hydraulic conductivity $K$ and by the hydraulic gradient $\Psi$.

$q = K\Psi$

#### Hydraulic conductivity $K$ ($m/s$)
The hydraulic conductivity caracterize the ability of water to move through a medium (ex: soil or rock) and is controlled by the porosity of the material $\kappa$ and the viscosity of the fluid $\mu$. It can be described with by $K=\kappa \frac{\rho g}{\mu}$, where $\rho$ is the density of water and $g$ the gravity.

Example of hydraulic conductivity for different type of material (in m/s):\
Sand	103 to 105\
Silt	105 to 108\
Clay	108 to 1012

#### Hydraulic gradient $\Psi$ ()
The hydraulic gradient controls controlled the force applied on the water column, which will force the water through the soil. The larger the water column, the water the water will move through the soil at the bottom of the water column. The hydraulic gradient is described with $\Psi = \frac{H}{d}$, where $H$ is the entire water column and $d$ is the soil saturated column.






Define Parameters for the infiltration:

1. Play with the hydraulic conductivity parameter K. Remember, it controls how fast the water can move through the soil. A larger K (ex: 1e-3) means that water will move faster and a smaller K (ex: 1e-6) means that the water will infiltrate slower. 

2. We determin initial values of $d$ and $h$.

Here, we have two different cases: 1 and 2.  

```python
#Define hydraulic conductivity parameters for the two different simulation the infiltration component
K = 1.0E-5
K2 = 1.0E-5

hydraulic_conductivity = mg.ones('node')*K
hydrualic_conductivity2 = mg2.ones('node')*K2

# Set initial values for h and d for both cases (VERIFY THAT THIS IS TRUE???)

## initial soil infilatration depth in the soil -- case 1
d = mg.add_ones("soil_water_infiltration__depth", at="node", dtype=float)
d = mg.at_node['soil_water_infiltration__depth']
d *= 0.2

## initial soil infilatration depth in the soil -- case 1
d2 = mg2.add_ones("soil_water_infiltration__depth", at="node", dtype=float)
d2 = mg2.at_node['soil_water_infiltration__depth']
d2 *= 0.2

## Surface water case 1
mg.add_zeros('surface_water__depth',at='node')  # a veneer of water  the model
h = mg.at_node['surface_water__depth']
h += 0.0001

## Surface water case 2
mg2.add_zeros('surface_water__depth',at='node')  # a veneer of water  the model
h2 = mg2.at_node['surface_water__depth']
h2 += 0.0001
```

```python
## Initialize model
SI = SoilInfiltrationGreenAmpt(mg,hydraulic_conductivity=hydraulic_conductivity)
of = OverlandFlow(mg, steep_slopes=True)

## Initialize model
SI2 = SoilInfiltrationGreenAmpt(mg2,hydraulic_conductivity=hydrualic_conductivity2)
of2 = OverlandFlow(mg2, steep_slopes=True)
```

```python
## Change storm times to minutes
interstorm_dts_min = np.asanyarray(interstorm_dts)*60 ## in minutes
storm_dts_min = np.asanyarray(storm_dts)*60 ## in minutes
global_time = np.sum(interstorm_dts_min) + np.sum(storm_dts_min)


interstorm_dts_min2 = np.asanyarray(interstorm_dts2)*60 ## in minutes
storm_dts_min2 = np.asanyarray(storm_dts2)*60 ## in minutes
global_time2 = np.sum(interstorm_dts_min2) + np.sum(storm_dts_min2)


## Arrays for saving inputs and counters -- case 1
SI_mean_vec = []
outlet_depth = []
outlet_times = []
elapsed_time = 0
node_of_max_q = 2126

## Arrays for saving inputs and counters -- case 2
SI_mean_vec2 = []
outlet_depth2 = []
outlet_times2 = []
elapsed_time2 = 0
node_of_max_q2 = 2126


## Better color map for plotting overlandflow
colors = [(0,0.2,1,i) for i in np.linspace(0,1,3)]
cmap = mcolors.LinearSegmentedColormap.from_list('mycmap', colors, N=10)
```

```python
## Main loop
for i in  range(0,len(storm_dts_min)):
    storm_t = storm_dts_min[i]
    interstorm_t = interstorm_dts_min[i]
    total_time_step = storm_t + interstorm_t
    storm_time = 0 
    print(str(np.round((elapsed_time/global_time)*100))+ '% from total run time')
    while storm_time < total_time_step:
        if storm_time < (total_time_step - interstorm_t):
            mg.at_node['surface_water__depth'] += (intensities[i]/(60))*0.001*10 ## meter in 10 minutes
            dt_overland = 10
            SI.run_one_step(dt=dt_overland)
            SI_mean_vec.append(np.mean(mg.at_node['soil_water_infiltration__depth']))
            of.run_one_step(dt=dt_overland)
            elapsed_time += dt_overland
            outlet_times.append(elapsed_time) 
            outlet_depth.append(mg.at_node['surface_water__depth'][node_of_max_q])
            storm_time += dt_overland
        else:
            SI.run_one_step(dt=dt_overland)
            of.run_one_step(dt=dt_overland)
            SI_mean_vec.append(np.mean(mg.at_node['soil_water_infiltration__depth']))
            # Increased elapsed time
            elapsed_time += dt_overland
            outlet_times.append(elapsed_time) 
            outlet_depth.append(mg.at_node['surface_water__depth'][node_of_max_q])
            storm_time += dt_overland 
        if elapsed_time % 1000 == 0:    
            fig = plt.figure(1)
            imshow_grid(mg,'topographic__elevation',colorbar_label='Elevation (m)')
            imshow_grid(mg,'surface_water__depth',limits=(0,0.6),cmap=cmap,colorbar_label='Water depth (m)')
            plt.title(f'Time = {round(elapsed_time,1)} s')
            plt.show()
```

```python
## Main loop
for i in  range(0,len(storm_dts_min2)):
    storm_t2 = storm_dts_min2[i]
    interstorm_t2 = interstorm_dts_min2[i]
    total_time_step2 = storm_t2 + interstorm_t2
    storm_time2 = 0 
    print(str(np.round((elapsed_time2/global_time2)*100))+ '% from total run time')
    while storm_time2 < total_time_step2:
        if storm_time2 < (total_time_step2 - interstorm_t2):
            mg2.at_node['surface_water__depth'] += (intensities2[i]/(60))*0.001*10 ## meter in 10 minutes
            dt_overland2 = 10
            SI2.run_one_step(dt=dt_overland2)
            SI_mean_vec2.append(np.mean(mg2.at_node['soil_water_infiltration__depth']))
            of2.run_one_step(dt=dt_overland2)
            elapsed_time2 += dt_overland2
            outlet_times2.append(elapsed_time2) 
            outlet_depth2.append(mg2.at_node['surface_water__depth'][node_of_max_q])
            storm_time2 += dt_overland2
        else:
            SI2.run_one_step(dt=dt_overland2)
            of2.run_one_step(dt=dt_overland2)
            SI_mean_vec2.append(np.mean(mg2.at_node['soil_water_infiltration__depth']))
            # Increased elapsed time
            elapsed_time2 += dt_overland2
            outlet_times2.append(elapsed_time2) 
            outlet_depth2.append(mg2.at_node['surface_water__depth'][node_of_max_q])
            storm_time2 += dt_overland2
        if elapsed_time % 1000 == 0:    
            fig2 = plt.figure(2)
            imshow_grid(mg2,'topographic__elevation',colorbar_label='Elevation (m)')
            imshow_grid(mg2,'surface_water__depth',limits=(0,0.6),cmap=cmap,colorbar_label='Water depth (m)')
            plt.title(f'Time = {round(elapsed_time2,1)} s')
            plt.show()
```
